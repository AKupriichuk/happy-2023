**Visualization of the world map according to the level of happiness 2023**

```python
import pandas as pd

df = pd.read_csv('/content/WHR2023.csv')
df.head()
```

Let's see if there are empty values

```python
print("Number of null or missing values in 2023 report: ",df.isnull().sum())
```

The column we need doesn't have them, so let's move on

```python
import plotly.io as pio
import plotly.express as px
fig4 = px.choropleth(data_frame=df,
                    locations=df['Country name'],
                    locationmode='country names',
                    color=df['Ladder score'],
                    color_continuous_scale='sunset',
                    projection = 'robinson',
                    title = 'The world map according to the level of happiness 2023'
                    )

fig4.show()
```

As we can see, the happiest people are in Finland, and the unhappiest are in Afghanistan
